```python
news=[]
for x in np.arange(126,300):#all bages in site
    
    url=f"https://www.youm7.com/Section/%D8%A3%D8%AE%D8%A8%D8%A7%D8%B1-%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6%D8%A9/298/{x}"
    req=requests.get(url)
    soup=BeautifulSoup(req.content,"html.parser") 
    jn=soup.find_all("p") 
    for i in jn: 
        
        try:
            news.append(i.text) 
        except: 
            news.append(np.nan)

    news.pop()
        
```
